```python
import pandas as pd
```

```python
mydata = pd.read_csv('./StudentPerformanceFactors.csv')

mydata
```

```python
mydata = mydata.dropna()
mydata
```

```python
mydata.head()
```

```python
mydata.tail()
```

```python
Student_Score_Features = ["Hours_Studied","Attendance","Sleep_Hours","Previous_Scores","Tutoring_Sessions","Physical_Activity"]
```

```python
y = mydata.Exam_Score
X = mydata[Student_Score_Features]
```

```python
X.describe()
```

```python
X.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

linear_regression = LinearRegression()

X_train, X_test, y_train, y_test = train_test_split(X,y,train_size= 0.8,random_state=10)


student_model = linear_regression.fit(X_train,y_train)

print(student_model)

print("Successfully created the model! \n Let's go test it!")
```

```python
prediction = student_model.predict(X_test)

print("Model Prediction In Action".center(60))

pred_data = {"Y_Predict":prediction,"Y_Actual":y_test}

result = pd.DataFrame(pred_data)

result
```

# Bonus Task

### Trying polynomial regression and comparing the performance

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score


degree = 2

poly = PolynomialFeatures(degree=degree,include_bias=False)
X_train_poly= poly.fit_transform(X_train)

model = LinearRegression()

model.fit(X_train_poly,y_train)

# Prediction
X_test_poly = poly.fit_transform(X_test)

prediction = model.predict(X_test_poly)

pred_data = {"Y_Predict":prediction,"Y_Actual":y_test}

result = pd.DataFrame(pred_data)

print(result)

# 5. Evaluate model

r2 = r2_score(y_train, model.predict(X_train_poly))
print(f"Polynomial Degree: {degree}")
print(f"Model Coefficients: {model.coef_}")
print(f"Model Intercept: {model.intercept_}")
print(f"R² Score: {r2:.4f}")
```
